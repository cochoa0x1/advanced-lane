```python
from os import listdir, path
import numpy as np
import cv2
import json
```

```python
%matplotlib inline
```

```python
def calculate_calibration_points(img):
    '''calculates the calibration image and corner points for a given input image file'''
    
    #convert image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    #we aren't sure how many corners are in the given image, so just try a few combinations until one works
    for nx in [9,8,7,6]:
        for ny in [6,5]:
            ret, corners = cv2.findChessboardCorners(gray, (nx,ny),None)
            
            if ret == False:
                continue
                
            object_points = np.array([ [j,i,0] for i in range(ny) for j in range(nx)], np.float32)
            
            return object_points, corners
        
def calibrate(im_path):
    '''performs camera calibration on a directory of images.
        returns object points, corners, and img shape.
    '''
    obj_points = []
    corner_points =[]
    
    for im_file in listdir(im_path):
        if im_file.split('.')[-1] not in ['jpeg','jpg','png']:
            print('skipping %s because file type is not supported'%im_file)
            continue
            
        print(im_file)
        
        #read in an image
        img = cv2.imread(path.join(im_path, im_file))
        #calculate the corners
        
        object_points, corners = calculate_calibration_points(img)
        
        #save the results away 
        if object_points is not None:
            obj_points.append(object_points)
            corner_points.append(corners)
        
    return obj_points, corner_points, img.shape[0:2]
        
```

```python
object_points,corners, img_shape = calibrate('../camera_cal/')
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(object_points, corners, img_shape,None,None)
```

```python
camera_cal = {'mtx':mtx.tolist(), 'dist':dist.tolist()}

file_name = 'camera.json'
with open(file_name, 'w') as f:
    f.write(json.dumps(camera_cal, indent=4))
    print('camera calibration saved to %s'%file_name)
```
