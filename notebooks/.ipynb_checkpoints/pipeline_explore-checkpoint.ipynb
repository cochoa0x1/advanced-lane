```python
import numpy as np
import cv2
import json
from os import listdir, path
%pylab inline
```

```python
def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def maybe_grayscale(img):
    #convert the image to gray scale if it isnt already single channel
    if len(img.shape) == 3 and img.shape[-1] !=1:
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    else:
        gray = img
    
    return gray
        

def apply_mask(x, thresh, norm=False):
    
    if norm:
        z = np.uint8(255 * x/np.max(x))
    else:
        z = x
        
    mask = np.zeros_like(z)
    mask[ (z >= thresh[0]) & (z <= thresh[1])]=1
    return mask
    
def abs_thresh(img, orient='x', sobel_kernel=3, thresh=(0, 255)):
    
    gray = maybe_grayscale(img)
    if orient=='x':
        s = np.abs(cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel))
    elif orient=='y':
        s = np.abs(cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel))
        
    return apply_mask(s,thresh, norm=True)
    
def mag_thresh(img, sobel_kernel=3, thresh=(0, 255)):
    
    gray = maybe_grayscale(img)
    sx = np.abs(cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel))
    sy = np.abs(cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel))
    
    #magnitude
    m = np.sqrt(sx**2 + sy**2)
    
    
    
    return apply_mask(m,thresh, norm=True)

def angle_thresh(img, sobel_kernel=3, thresh=(0, np.pi/2)):
    gray = maybe_grayscale(img)
    sx = np.abs(cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel))
    sy = np.abs(cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel))
    
    theta = np.arctan2(sy,sx)
    return apply_mask(theta,thresh, norm=False)
```

```python
#undistort
undist = cv2.undistort(im, mtx, dist, None, mtx)
imshow(undist)
```

```python
sky_y = 450
undist = undist[sky_y:,:,:]
imshow(undist)
```

```python
#trapezoid mask
ys, xs, _ = undist.shape


points = np.array([[210,ys],[595,0],[685,0],[1100,ys]], np.float32)
dst = np.array([[210,4*ys],[210,0],[1070,0],[1070,4*ys]], np.float32)

can_region=region_of_interest(undist,[np.int32(points)])
imshow(can_region)

#plot([640],[418],'*')
figure()

M = cv2.getPerspectiveTransform(points, dst)

warped = cv2.warpPerspective(undist, M, (xs,4*ys))
imshow(warped)
```

```python
#HLS it
hls = cv2.cvtColor(warped, cv2.COLOR_RGB2HLS)

#keep the S channel only
S = hls[:,:,2]
imshow(S, cmap='gray')
colorbar()
H = hls[:,:,1]
figure()
imshow(H, cmap='gray')
colorbar()

#S = gaussian_blur(S,5)
#H = gaussian_blur(H,5)
```

```python
def exp_thresh(x, target, beta):
    return 255*np.exp(-1.0*beta*np.square((x-1.0*target)))
    
```
