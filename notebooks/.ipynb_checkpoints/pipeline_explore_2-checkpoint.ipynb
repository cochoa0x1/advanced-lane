```python
import numpy as np
import cv2
import json
from os import listdir, path
from scipy.ndimage.filters import gaussian_filter
from scipy.optimize import minimize
%pylab inline
```

```python
def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def maybe_grayscale(img):
    #convert the image to gray scale if it isnt already single channel
    if len(img.shape) == 3 and img.shape[-1] !=1:
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    else:
        gray = img
    
    return gray
        

def apply_mask(x, thresh, norm=False):
    
    if norm:
        z = np.uint8(255 * x/np.max(x))
    else:
        z = x
        
    mask = np.zeros_like(z)
    mask[ (z >= thresh[0]) & (z <= thresh[1])]=1
    return mask
    
def abs_thresh(img, orient='x', sobel_kernel=3, thresh=(0, 255)):
    
    gray = maybe_grayscale(img)
    if orient=='x':
        s = np.abs(cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel))
    elif orient=='y':
        s = np.abs(cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel))
        
    return apply_mask(s,thresh, norm=True)
    
def mag_thresh(img, sobel_kernel=3, thresh=(0, 255)):
    
    gray = maybe_grayscale(img)
    sx = np.abs(cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel))
    sy = np.abs(cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel))
    
    #magnitude
    m = np.sqrt(sx**2 + sy**2)
    
    return apply_mask(m,thresh, norm=True)

def angle_thresh(img, sobel_kernel=3, thresh=(0, np.pi/2)):
    gray = maybe_grayscale(img)
    sx = np.abs(cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel))
    sy = np.abs(cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel))
    
    theta = np.arctan2(sy,sx)
    return apply_mask(theta,thresh, norm=False)

def filter_by_color(img):
    
    #HLS it
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)

    S = hls[:,:,2]
    L = hls[:,:,1]
    
    a=apply_mask(S,(190,255))
    b=apply_mask(L,(220,255))
    
    return np.maximum(a,b)
    
    
def filter_by_geometry(img):
    a = abs_thresh(img,'x', sobel_kernel=5, thresh=(30,255))
    a = angle_thresh(a, sobel_kernel=5, thresh=(.75,np.pi/2))
    return a


def warp(img):
    ys, xs = img.shape[:2]
    

    points = np.array([[210,ys],[595,450],[685,450],[1100,ys]], np.float32)
    dst = np.array([[210,ys],[210,0],[1070,0],[1070,ys]], np.float32)


    M = cv2.getPerspectiveTransform(points, dst)

    warped = cv2.warpPerspective(img, M, (xs,ys))
    
    return warped

def unwarp(img):
    ys, xs = img.shape[:2]
    

    points = np.array([[210,ys],[595,450],[685,450],[1100,ys]], np.float32)
    dst = np.array([[210,ys],[210,0],[1070,0],[1070,ys]], np.float32)


    M = cv2.getPerspectiveTransform(dst,points)

    warped = cv2.warpPerspective(img, M, (xs,ys))
    
    return warped


def lane_lines(img):
    
    undist = cv2.undistort(img, mtx, dist, None, mtx)
    warped = warp(undist)
    
    a = filter_by_color(warped)
    b = filter_by_geometry(warped)

    return np.maximum(a,b)
```

```python
def gauss(x,mu, sig):
    z=  np.exp( -1.0*(x-mu)**2  / (2*sig**2))
    return z / np.sum(z)

def naive_prior(step):
    w,y = 640,720 #shape
    ys = np.arange(0,y,step)
    x=np.arange(w)
    return np.array([ gauss(x,mu=220,sig=50+.25*i) for i in ys])

def observation_dist(img,y,high,low):
    '''calculates a histogram for a slice of lane line image'''
    #take a slice of image and calculate a histogram
    p=np.sum(img[ max(y-low,0):y+high,:], axis=0)
    
    #if somehow we have no points in this slice, then revert to uniform dist
    if sum(p) ==0.0:
        p = np.ones(len(p))
        #print('fail',y)
        
    #make it a bit smooth
    p = gaussian_filter(p, sigma=5)
    
    #normalize
    return p / np.sum(p)
    
#movement left/right per frame
p_new_frame = gauss(np.arange(-200,200+1,1),mu=0, sig = 20)

#movement left/right per increase in y
p_move=gauss(np.arange(-100,100+1,1),mu=0, sig = 1)

def line_prob(img,prior_last_frame,p_new_frame,p_move,step=5):

    P=[]

    for i,y in enumerate(np.arange(0,720,step)):

        #prior frame
        p_last_frame = prior_last_frame[i,:]
        
        #fuzzy it cause we moved
        p_last_frame=np.convolve(p_last_frame,p_new_frame, mode='same')
        #p_last_frame/=np.sum(p_last_frame)

        #get our current observation distribution
        p = observation_dist(img,y,50,50)
        #print(y)
        #update. if this is our first frame then our prior is just the prior frame
        if i==0:
            prior=p_last_frame
        else:
            prior = P[-1]
            #move left or right a bit
            prior = np.convolve(prior,p_move, mode='same')
            #mix with last frame
            prior*=p_last_frame

        #update with our observations
        z=p*prior
        if sum(z)==0:
            z=prior
        z /= np.sum(z)
        P.append(z)
        
    return np.array(P)


ym_per_pix = 30/720 # meters per pixel in y dimension
xm_per_pix = 3.7/700 # meters per pixel in x dimension

def fit(left,right):
    
    def get_points(Z):
        x,y,w=[],[],[]
        for i in np.arange(720):
            for j in np.arange(640):
                _w = Z[i,j]
                if _w > 0:
                    x.append(j)
                    y.append(i)
                    w.append(_w)

        return xm_per_pix*np.array(x) ,ym_per_pix*np.array(y) ,np.array(w)
    
    
    xl,yl,wl =get_points(left)
    xr,yr,wr =get_points(right)
    xr+=640*xm_per_pix
    
    

    #our error function
    def f(p):
        A,B,C,D = p
        #D=850
        pred_x_l = A*(yl**2)+B*yl+C
        pred_x_r = A*(yr**2)+B*yr+(C+D)
        return np.sum(wl*(xl-pred_x_l)**2)/sum(wl)+ np.sum(wr*(xr-pred_x_r)**2)/sum(wr)
    
    p=[0,0,220*xm_per_pix,885*xm_per_pix]
    bounds = [(None,None),(None,None),(None,None),(870.0*xm_per_pix,900.0*xm_per_pix)]
    fit = minimize(f,p,method='L-BFGS-B',bounds=bounds)
    
    return fit.x
```

```python
for img_file in listdir('../test_images/'):
    im = cv2.imread(path.join('../test_images/',img_file))
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    c = lane_lines(im)
    
    figure(figsize=(12,4))
    subplot(1,4,1)
    imshow(warp(im))
    
    plot(-1*np.sum(c[int(c.shape[0]/2):,:],axis=0)+(c.shape[0]), color='red', linewidth=2)
    xlim(0,c.shape[1])
    ylim(c.shape[0],0)
    
    subplot(1,4,2)
    imshow(c[:,:],cmap='gray')
    
    #subplot(1,3,3)
    
    #left/right split
    L = c[:,: int(c.shape[1]/2)]
    R = c[:,int(c.shape[1]/2):]
    
    left = L[::-1,:]
    right = R[::-1,:]

    naive_last_frame_right = naive_prior(step=1)
    naive_last_frame_left = naive_last_frame_right[:,::-1]

    P_left=line_prob(left,naive_last_frame_right,p_new_frame,p_move,step=1)
    P_right=line_prob(right,naive_last_frame_left,p_new_frame,p_move,step=1)
    
    subplot(1,4,3)
    r =fit(left*P_left,right*P_right)
    print(r)
    
    y=np.arange(720)*ym_per_pix
    left_fit = (r[0]*y**2+r[1]*y+r[2])/xm_per_pix
    right_fit = (r[0]*y**2+r[1]*y+r[2]+r[3])/xm_per_pix
    
    y/=ym_per_pix

    imshow(warp(im)[::-1,:])
    plot(left_fit,y)
    plot(right_fit,y)

    ylim(0,720)
    xlim(0,1280)
    
    y_eval=0
    r_curve =  ((1 + (2*r[0]*y_eval + r[1])**2)**1.5) / np.absolute(2*r[0])
    lane_center = (r[2]+r[3]*.5)
    position =  c.shape[1]*xm_per_pix*.5 -  lane_center
    
    subplot(1,4,4)
    
    
    warp_zero = np.zeros_like(c).astype(np.uint8)
    color_warp = np.dstack((warp_zero, warp_zero, warp_zero))
    
    
    # Recast the x and y points into usable format for cv2.fillPoly()
    pts_left = np.array([np.transpose(np.vstack([left_fit, y]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fit, y])))])
    pts = np.hstack((pts_left, pts_right))

    # Draw the lane onto the warped blank image
    cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))

    color_warp =color_warp[::-1,:]
    # Warp the blank back to original image space using inverse perspective matrix (Minv)
    newwarp = unwarp(color_warp)
    
    undist = cv2.undistort(im, mtx, dist, None, mtx)
    # Combine the result with the original image
    result = cv2.addWeighted(undist, 1, newwarp, 0.3, 0)
    imshow(result)
    
    plot([lane_center/xm_per_pix,lane_center/xm_per_pix],[c.shape[0],c.shape[0]-100])
    
    plot([c.shape[1]*.5]*2,[c.shape[0],c.shape[0]-200])
    
    title([r_curve,position])
    xlim(0,c.shape[1])
    ylim(c.shape[0],0)


    
    tight_layout()
    
```
